This Jupyter notebook modifies the csv file and store it as a new csv for further use as my dataset is pretty big.

```python
import pandas as pd
```

```python
df = pd.read_csv("data/311_Service_Requests_20250130.csv")
```

```python
columns = [
    'CREATED_DEPARTMENT', 'OWNER_DEPARTMENT', 'CREATED_DATE', 'CLOSED_DATE', 
    'STREET_ADDRESS', 'ZIP_CODE', 'COMMUNITY_AREA', 'CREATED_HOUR', 
    'CREATED_DAY_OF_WEEK', 'CREATED_MONTH', 'LATITUDE', 'LONGITUDE', 'LOCATION'
]

df_selected = df[columns]
df_selected.head()
print(df_selected.shape)
```

```python
df_selected_cleaned = df_selected.dropna()
print(df_selected_cleaned.shape)
```

```python
df_selected_cleaned['CREATED_DATE'] = pd.to_datetime(df_selected_cleaned['CREATED_DATE'], errors='coerce')

df_2021 = df_selected_cleaned[df_selected_cleaned['CREATED_DATE'].dt.year >= 2021]
```

```python
df_2021.head()
print(df_2021.shape)
```

```python
df_2021.to_csv('data/311_Service_Requests_2021_onwards.csv', index=False)
```
